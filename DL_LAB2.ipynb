```python
# STEP 1: Given data
x = [0.75, 0.65, 0.80]     # input vector
w = [0.2, 0.4, 0.6]       # weight vector

b = -0.5              # bias
n = 0.1                # learning rate
y = 1                    # target output


z = 0
for i in range(len(x)):
    z = z + (x[i] * w[i])
z = z + b

if z >= 0:
    y_pred = 1
else:
    y_pred = 0

error = y - y_pred

for i in range(len(w)):
    w[i] = w[i] + n * error * x[i]

b = b + n * error

print("Input vector x :", x)
print("Net input z    :", z)
print("Predicted y   :", y_pred)
print("Actual y      :", y)
print("Error         :", error)
print("Updated w     :", w)
print("Updated b     :", b)
```
